### Função de ativação sigmoide

$$f(x) = \frac{1}{1 + e^{-\beta \cdot x}}$$

- Implemente uma função em Python que retorne o valor da função sigmóide para uma dada entrada
- Nesta função, considere beta como parâmetro nomeado com valor igual a 1.0
- Plote o gráfico da função no intervalo de entrada previamente especificado
- Implemente uma função que retorne o valor da derivada da função sigmoide
- Em um mesmo gráfico, ilustre 3 valores distintos para $\beta$:
  - $\beta = 0.7$ na cor laranja
  - $\beta = 0.5$ na cor verde
  - $\beta = 0.3$ na cor vermelha
  - Use legendas no gráfico

```python
# 1 - Implementar a função sigmoide
def sigmoid(x, beta=1.0):
    return 1 / (1 + np.exp(-beta * x))
```

```python
# 2 - Usar beta como parâmetro nomeado com valor 1.0
print(sigmoid(0))   # esperado 0.5
print(sigmoid(2))   # valor > 0.5
```

```python
# 3 - Plotar o gráfico da função sigmoide no intervalo de entrada especificado
plt.figure(figsize=(6,4))
plt.plot(x, sigmoid(x), label="Sigmoid β=1.0", color="blue")
plt.title("Função Sigmoide (β=1.0)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 4 - Implementar a derivada da função sigmoide
def sigmoid_derivative(x, beta=1.0):
    s = sigmoid(x, beta)
    return beta * s * (1 - s)
```

```python
# Extra - Plotar o gráfico da derivada da função sigmoide (β=1.0)
plt.figure(figsize=(6,4))
plt.plot(x, sigmoid_derivative(x), label="Sigmoid Derivada β=1.0", color="red")
plt.title("Derivada da Função Sigmoide (β=1.0)")
plt.xlabel("x")
plt.ylabel("f'(x)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 5 - Em um mesmo gráfico, ilustrar 3 valores distintos de beta (0.7, 0.5, 0.3) com cores e legendas
plt.figure(figsize=(6,4))
plt.plot(x, sigmoid(x, beta=0.7), color="orange", label="β=0.7")
plt.plot(x, sigmoid(x, beta=0.5), color="green", label="β=0.5")
plt.plot(x, sigmoid(x, beta=0.3), color="red", label="β=0.3")
plt.title("Função Sigmoide com diferentes valores de β")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend(loc="lower center")
plt.show()
```

### Função de Ativação ReLU (Unidade Linear Retificada)

$$f(x) = max(0,x)$$

- Implemente uma função em Python que retorne o valor da função ReLU para uma dada entrada
- Plote o gráfico da função no intervalo de entrada previamente especificado
- Implemente uma função que retorne o valor da derivada da função ReLU

```python
# 1 - Implementar a função ReLU
def relu(x):
    return np.maximum(0, x)
```

```python
# 2 - Plotar o gráfico da função ReLU no intervalo de entrada especificado
plt.figure(figsize=(6,4))
plt.plot(x, relu(x), label="ReLU", color="blue")
plt.title("Função de Ativação ReLU")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 3 - Implementar a função derivada da ReLU
def relu_derivative(x):
    return np.where(x > 0, 1, 0)
```

```python
# Extra - Plotar o gráfico da derivada da ReLU no intervalo de entrada
plt.figure(figsize=(6,4))
plt.plot(x, relu_derivative(x), label="ReLU Derivada", color="red")
plt.title("Derivada da Função ReLU")
plt.xlabel("x")
plt.ylabel("f'(x)")
plt.grid(True)
plt.legend()
plt.show()
```

### Função de ativação Tangente Hiperbólica

$$f(x) = tanh(x)$$

- Implemente uma função em Python que retorne o valor da função tangente hiperbólica para uma dada entrada
- Plote o gráfico da função no intervalo de entrada previamente especificado
- Implemente uma função que retorne o valor da derivada da função em questão
- Utilize bibliotecas para auxiliar nesta tarefa
- Considere o intervalo (-1,1)

```python
# 1 - Implementar a função tangente hiperbólica
def tanh(x):
    return np.tanh(x)
```

```python
# 2 - Plotar o gráfico da função tangente hiperbólica no intervalo de entrada
plt.figure(figsize=(6,4))
plt.plot(x, tanh(x), label="tanh(x)", color="blue")
plt.title("Função de Ativação Tangente Hiperbólica")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 3 - Implementar a função derivada da tangente hiperbólica
def tanh_derivative(x):
    return 1 - np.tanh(x)**2
```

```python
# Extra - Plotar o gráfico da derivada da tangente hiperbólica no intervalo (-1, 1)
x_interval = np.arange(-1, 1.1, 0.1)

plt.figure(figsize=(6,4))
plt.plot(x_interval, tanh_derivative(x_interval), label="tanh'(x)", color="red")
plt.title("Derivada da Função Tangente Hiperbólica (intervalo -1,1)")
plt.xlabel("x")
plt.ylabel("f'(x)")
plt.grid(True)
plt.legend()
plt.show()
```

### Função de Ativação Gaussiana

$$f(x) = \exp \left( \frac{-(x - \mu)^2}{2\sigma^2} \right)$$

- $\mu$ é o ponto médio (centro) da curva
- $\sigma$ denota o raio de abertura
- Implemente uma função em Python que retorne o valor da função gaussiana para uma dada entrada
- Plote o gráfico da função no intervalo de entrada previamente especificado
- Implemente uma função que retorne o valor da derivada da função em questão


Considere também os seguintes valores ao plotar a função

- Parâmetros nomeados $\mu = 0$ e $\sigma = 1$
- $\mu = 0, \sigma = 0.3$
- $\mu = 0, \sigma = 0.5$
- $\mu = 0, \sigma = 0.7$
- $\mu = 0, \sigma = 0.9$
- $\mu = 0, \sigma = 1.1$

```python
# 1 - Implementar a função gaussiana
def gaussian(x, mu=0, sigma=1):
    return np.exp(-((x - mu)**2) / (2 * sigma**2))
```

```python
# 2 - Plotar o gráfico da função gaussiana no intervalo de entrada (mu=0, sigma=1 por padrão)
plt.figure(figsize=(6,4))
plt.plot(x, gaussian(x), label="μ=0, σ=1", color="blue")
plt.title("Função de Ativação Gaussiana (padrão μ=0, σ=1)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 3 - Implementar a derivada da função gaussiana
def gaussian_derivative(x, mu=0, sigma=1):
    return -(x - mu) / (sigma**2) * gaussian(x, mu, sigma)
```

```python
# Extra - Plotar o gráfico da derivada da função gaussiana para σ=1
plt.figure(figsize=(6,4))
plt.plot(x, gaussian_derivative(x, mu=0, sigma=1), label="Derivada μ=0, σ=1", color="red")
plt.title("Derivada da Função Gaussiana (μ=0, σ=1)")
plt.xlabel("x")
plt.ylabel("f'(x)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 4 - Plotar o gráfico da função gaussiana para diferentes valores de sigma (μ=0 fixo)
sigmas = [0.3, 0.5, 0.7, 0.9, 1.1]

plt.figure(figsize=(8,6))
for s in sigmas:
    plt.plot(x, gaussian(x, mu=0, sigma=s), label=f"μ=0, σ={s}")
    
plt.title("Função Gaussiana para diferentes valores de σ (μ=0)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```

### Função de Ativação Leaky ReLU

$$f(x) = \left\{\begin{array}{cc} 0.01\cdot x & x < 0\\
x & x \geq 0
\end{array}
\right.$$

- Implemente uma função em Python que retorne o valor da função Leaky ReLU para uma dada entrada
- Plote o gráfico da função no intervalo de entrada previamente especificado
- Implemente uma função que retorne o valor da derivada da função Leaky ReLU (consulte a literatura!)

```python
# 1 - Implementar a função Leaky ReLU
def leaky_relu(x, alpha=0.01):
    return np.where(x >= 0, x, alpha * x)
```

```python
# 2 - Plotar o gráfico da função Leaky ReLU no intervalo de entrada
plt.figure(figsize=(6,4))
plt.plot(x, leaky_relu(x), label="Leaky ReLU (α=0.01)", color="blue")
plt.title("Função de Ativação Leaky ReLU")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 3 - Implementar a função derivada da Leaky ReLU
def leaky_relu_derivative(x, alpha=0.01):
    return np.where(x >= 0, 1, alpha)
```

```python
# Extra - Plotar o gráfico da derivada da Leaky ReLU no intervalo de entrada
plt.figure(figsize=(6,4))
plt.plot(x, leaky_relu_derivative(x), label="Leaky ReLU Derivada (α=0.01)", color="red")
plt.title("Derivada da Função Leaky ReLU")
plt.xlabel("x")
plt.ylabel("f'(x)")
plt.grid(True)
plt.legend()
plt.show()
```
